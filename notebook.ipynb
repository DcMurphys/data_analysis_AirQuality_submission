### Cleaning Data

Rename kolom 'PM2.5' menjadi kolom 'PM2p5'

```python
all_stations_df.rename(columns = {'PM2.5' : 'PM2p5'}, inplace=True)
```

Mengisi data yang memiliki missing value (NaN) dengan nilai terdekat menggunakan metode interpolate 'nearest'

```python
all_stations_df_temp = all_stations_df.interpolate(method='nearest')
```

Menghilangkan kolom 'TEMP', 'PRES',	'DEWP',	'RAIN',	'WSPM', dan 'wd' karena bukan merupakan indikator yang akan diuji dalam analisis data berdasarkan pertanyaan bisnis yang telah diajukan

```python
all_stations_df_temp = all_stations_df_temp.drop(columns=['TEMP','PRES','DEWP','WSPM','RAIN','wd'])
all_stations_df_temp.info()
```

Menghilangkan data outlier dengan menggunakan pendekatan interkuartil pada dataset

```python
cols = ['PM2p5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']
Q1 = all_stations_df_temp[cols].quantile(0.25)
Q3 = all_stations_df_temp[cols].quantile(0.75)
IQR = Q3 - Q1

condition = ~((all_stations_df_temp[cols] < (Q1 - 1.5 * IQR))).any(axis=1)

all_stations_df_clean = all_stations_df_temp[condition]
```

Pengecekan kembali apakah missing value sudah diisi atau tidak

```python
all_stations_df_clean.isna().sum()
```

Memformat ulang 'year', 'month', 'date', dan 'hour' ke dalam satu kolom bernama 'record_time', kemudian hasilnya disimpan sebagai dataframe 'all_stations_df_clean'

```python
all_stations_df_clean['record_time'] = pd.to_datetime(all_stations_df_clean[['year','month','day','hour']])
all_stations_df_clean = all_stations_df_clean.drop(columns=['year','month','day','hour'])
```

**Insight:**
- Missing value (NaN) pada beberapa parameter data, seperti 'PM2.5', 'PM10', 'SO2', dsb. diisi dengan nilai terdekat dengan menggunakan antara metode interpolasi 'nearest' karena parameter-parameter ini berisikan data kontinu.
- Data outlier yang ada pada dataset 'all_stations_df_temp' kemudian dihilangkan dengan menggunakan metode interkuartil, kemudian dataset tersebut diubah menjadi hanya menampilkan data yang nilainya memenuhi syarat interkuartil.
- Kolom berkaitan waktu pengamatan seperti 'year', 'month', 'day', 'hour' disatukan ke dalam sebuah kolom baru bernama 'record_time' dengan tipe data 'datetime' dengan menggunakan kombinasi teknik merge columns dengan 'pd.to_datetime' dan disimpan sebagai dataframe 'all_stations_df_temp'. Keempat kolom tersebut kemudian dihapus dengan menggunakan teknik drop, lalu hasilnya disimpan sebagai dataframe 'all_stations_df_clean'.  

## Exploratory Data Analysis (EDA)

### Explore Air Quality Index (AQI) based on Stations and its parameters (particulates)

Memeriksa parameter statistik dari dataset gabungan 'all_stations_df_clean' menggunakan teknik describe dengan menyertakan parameter include="all"

```python
all_stations_df_clean.describe(include="all")
```

Mencari nilai AQI berdasarkan nilai dari parameter pengukur yang diberikan pada dataset

```python
# Memeriksa kelengkapan nilai pada masing-masing variabel pengukur AQI
all_stations_df_clean['checks'] = (all_stations_df_clean['PM2p5'] > 0).astype(int) + \
                                (all_stations_df_clean['PM10'] > 0).astype(int) + \
                                (all_stations_df_clean['SO2'] > 0).astype(int) + \
                                (all_stations_df_clean['NO2'] > 0).astype(int) + \
                                (all_stations_df_clean['CO'] > 0).astype(int) + \
                                (all_stations_df_clean['O3'] > 0).astype(int)

# Menghitung nilai AQI dan memasukkannya ke dalam kolom baru bernama 'AQI_result'
all_stations_df_clean["AQI_result"] = round(all_stations_df_clean[["PM2p5", "PM10", "SO2", "NO2",
                                      "CO", "O3"]].median(axis = 1))
all_stations_df_clean.loc[all_stations_df_clean["PM2p5"] + all_stations_df_clean["PM10"] <= 0, "AQI_result"] = np.NaN
all_stations_df_clean.loc[all_stations_df_clean.checks < 3, "AQI_result"] = np.NaN
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana kadar polutan PM2.5 dan PM10 yang tercatat di semua stasiun pengamatan selama tahun 2013 - 2017?

Resampling data 'record_time' menjadi bulanan menggunakan teknik 'resample', lalu memformat label hasil resampling menjadi "%Y-%m".

```python
# Resampling data 'record_time' menjadi bulanan menggunakan teknik 'resample', lalu memformat label hasil resampling menjadi "%Y-%m".
monthly_record_df = all_stations_df_clean.resample(rule='M', on='record_time').agg({
    "station" : "nunique",
    "PM2p5" : "median",
    "PM10" : "median"
})

monthly_record_df.index = monthly_record_df.index.strftime('%Y-%m')
monthly_record_df = monthly_record_df.reset_index()
monthly_record_df.head()
```

Visualisasi data bulanan dalam bentuk line chart

```python
# Visualisasi data gabungan PM2.5 dan PM10 menggunakan line chart
columns_plot = ['PM2p5','PM10']

fig, ax = plt.subplots(figsize=(15,5))
ax.yaxis.set_major_formatter(ticker.EngFormatter())
for each in columns_plot:
    sns.lineplot(data=monthly_record_df, x='record_time', y=each, label=str(each), errorbar=None)
plt.legend(loc='upper left', labels=['PM2.5','PM10'])
plt.tight_layout()
plt.tick_params(axis='x', rotation=45)
plt.title("PM2.5 and PM10 Particle (Month-on-Month)", loc="center", fontsize=15)
plt.xlabel('Time of recording')
plt.ylabel('Particulate Matter Score in g/m3')
plt.show()
```

```python
# Visualisasi data PM2.5 dan PM10 secara berbeda menggunakan line chart
'''PM2.5'''
plt.figure(figsize=(18,5))
plt.title("PM2.5 Particle (Month-on-Month)", loc="center", fontsize=15)
plt.xlabel('Time of recording')
plt.ylabel('PM2.5 in g/m3')
plt.tick_params(axis ='x', rotation=45)
sns.lineplot(data=monthly_record_df,x='record_time',y='PM2p5')

'''PM10'''
plt.figure(figsize=(18,5))
plt.title("PM10 Particle (Month-on-Month)", loc="center", fontsize=15)
plt.xlabel('Time of recording')
plt.ylabel('PM10 in g/m3')
plt.tick_params(axis ='x', rotation=45)
sns.lineplot(data=monthly_record_df,x='record_time',y='PM10')
```

**Insight:**
- Visualisasi line-chart digunakan untuk mengilustrasikan pergerakan indeks PM2.5 dan PM10 yang dicatatkan di 12 stasiun pengamatan selama periode perekaman. Sampel perekaman ini kita ubah terlebih dahulu melalui teknik data resampling dari sampel yang direkam setiap jam ke bulanan.
- Dari visualisasi di atas, indeks polutan PM2.5 dan PM10 di 12 stasiun pengamatan biasanya akan meningkat ketika memasuki kuartal keempat setiap tahunnya hingga kuartal pertama tahun berikutnya. Hal ini dibuktikan dengan kenaikan kedua indeks polutan di 12 stasiun pengamatan, masing-masing tercatat pada Okt 2013 - Mei 2014, Okt 2014 - Mei 2015, dan Ags 2015 - Jan 2016.
- Pada gambar yang sama, indeks polutan PM2.5 dan PM10 justru merosot tepat ketika memasuki akhir tahun 2016.

### Pertanyaan 2: Bagaimana indeks kualitas udara (AQI) yang tercatat di semua stasiun pengamatan selama tahun 2013 - 2017?

Visualisasi data AQI bulanan (month-to-month) dalam bentuk line chart

```python
# Resampling data 'record_time' menjadi bulanan menggunakan teknik 'resample', lalu memformat label hasil resampling menjadi "%Y-%m".
monthly_record_df = all_stations_df_clean.resample(rule='M', on='record_time').agg({
    "station" : "nunique",
    "AQI_result" : "median",
})

monthly_record_df.index = monthly_record_df.index.strftime('%Y-%m')
monthly_record_df = monthly_record_df.reset_index()

# Visualisasi data AQI month-to-month
fig, ax = plt.subplots(figsize=(15,5))
ax.tick_params(axis='x', rotation=45)
ax.plot(monthly_record_df['record_time'], monthly_record_df['AQI_result'])
ax.set_title("Air Quality Index (AQI) (Month-on-Month)", loc="center", fontsize=15)
ax.set_xlabel('Time of recording')
ax.set_ylabel('AQI score')
plt.show()

# Visualisasi data AQI paling maksimum secara year-to-year

yearly_record_df = all_stations_df_clean.resample(rule='Y', on='record_time').agg({
    "station" : "nunique",
    "AQI_result" : "max",
})

yearly_record_df.index = yearly_record_df.index.strftime('%Y')
yearly_record_df = yearly_record_df.reset_index()

fig, ax = plt.subplots(figsize=(12,4))

sns.barplot(x="AQI_result", y="record_time", data=yearly_record_df.head())

ax.set_ylabel(None)
ax.set_xlabel(None)
plt.tight_layout()
ax.set_title("Highest Record of AQI Score (Year-on-Year)", loc="center", fontsize=15)
ax.tick_params(axis ='y', labelsize=12)
```

**Insight:**
- Visualisasi line-chart digunakan untuk mengilustrasikan pergerakan skor AQI yang dicatatkan di 12 stasiun pengamatan selama periode perekaman. Sampel perekaman ini kita ubah terlebih dahulu melalui teknik data resampling dari sampel yang direkam setiap jam ke bulanan.
- Skor Indeks kualitas udara (AQI) yang dicatatkan di 12 stasiun pengamatan mengalami lonjakan pada November 2013 - April 2014, Mei - Agustus 2014, November 2014 - Januari 2015, Agustus 2015 - Januari 2016, dan September 2016 - Januari 2017.
- Data ini sulit untuk dicarikan polanya secara pasti, namun kemungkinan besar lonjakan skor AQI akan terjadi paling cepat mulai bulan Agustus hingga paling lambat Maret atau April tahun berikutnya.
- Visualisasi bar-chart digunakan untuk membandingkan skor AQI tertinggi dari tahun ke tahun (YoY). Visualisasi ini menunjukkan bahwa tahun 2016 merupakan tahun dengan rekor skor AQI paling tinggi yang pernah dicatatkan sepanjang periode pencatatan di 12 stasiun pengamatan.

### Pertanyaan 3: Stasiun pengamatan mana yang mencatatkan nilai AQI paling baik dan paling buruk sepanjang periode perekaman?

**Insight:**

### Pertanyaan 5: Bagaimana pengelompokkan (simple clustering) sampel data ke dalam status terkait indeks kualitas udara (AQI) sesuai pedoman yang berlaku di China?

Mendefinisikan fungsi untuk menentukan status indeks kualitas udara (AQI) berdasarkan pedoman yang berlaku di China dengan menggunakan fungsi "get_AQI_status".

Berikut merupakan klasifikasi status AQI berdasarkan pedoman yang berlaku di China.
- Excellent: AQI <= 50
- Good: AQI <= 100
- Lightly Polluted: AQI <= 150
- Moderately Polluted: AQI <= 200
- Heavily Polluted: AQI <= 300
- Severly Polluted: AQI > 300

```python
def get_AQI_status(x):
    if x <= 50:
      return 'Excellent'
    elif x <= 100:
      return 'Good'
    elif x <= 150:
      return 'Lightly Polluted'
    elif x <= 200:
      return 'Moderately Polluted'
    elif x <= 300:
      return 'Heavily Polluted'
    elif x > 300:
      return 'Severely Polluted'
    else:
      return np.NaN
```

Klasifikasi (clustering) sampel rekaman secara sederhana (manual grouping) berdasarkan status indeks kualitas udara yang ditetapkan dalam pedoman yang berlaku di China, dengan mencocokkan indeks kualitas udara (AQI) dengan status yang didefinisikan pada fungsi 'get_AQI_status'

```python
all_stations_df_clean["AQI_status"] = all_stations_df_clean["AQI_result"].apply(lambda x: get_AQI_status(x))
all_stations_df_clean[~all_stations_df_clean.AQI_result.isna()].head(50)
```

```python
# Klasifikasi data ke dalam 'AQI_status' tertentu menggunakan conditioning dalam fungsi 'get_AQI_status'
all_stations_df_clean["AQI_status"] = all_stations_df_clean["AQI_result"].apply(lambda x: get_AQI_status(x))
all_stations_df_clean[~all_stations_df_clean.AQI_result.isna()].head(50)

# Pembuatan dataframe baru bernama 'AQI_status' untuk menangani data hasil clustering
all_stations_df_clean = all_stations_df_clean[~all_stations_df_clean.AQI_result.isna()]
AQI_status = all_stations_df_clean['AQI_status'].value_counts().reset_index()

# Visualisasi hasil clustering sederhana dalam bentuk jumlah sample dengan status tertentu menggunakan bar-chart
sns.barplot(data=AQI_status, x='AQI_status', y='count')
plt.tight_layout()
plt.title("AQI Status Classification")
plt.ylabel("n-count")
plt.tick_params(axis='x', rotation=45)
```

Penyajian korelasi AQI dengan indeks PM2.5 dan PM10 lengkap dengan pelabelan berdasarkan status keparahan polusi 'AQI_status' menggunakan scatter plot diagram
